# Q1
Some values in the the FlightNumber column are missing. These numbers are
meant to increase by 10 with each row so 10055 and 10075 need to be put in
place. Fill in these missing numbers and make the column an integer column
(instead of a float column).

```python
df1
```

```python
# 1. Converting all the elements in the series to a list
lst = [int(x) if str(x).isnumeric() else str(x) for x in list(df1['FlightNumber'])]
lst
```

```python
# 2. Loop to check if any elemnt is nan and replace it with the previous member + 10 and create a newlist
i = 0
while i + 1 < len(lst):
    if lst[i+1] == 'nan':
        lst[i+1] = float(lst[i]) + 10
    i += 1
lst
```

```python
# 3. Replacing the FlightNumber series in the real dataset with the above got list data
df1['FlightNumber'] = pd.Series(lst,name='Flight_Number')
```

```python
# 4. Typecasting to float and converting again to integer to avoid .0
df1['FlightNumber'] = df1['FlightNumber'].astype(float).astype(int)
df1
```

# Q2
The From_To column would be better as two separate columns! Split each
string on the underscore delimiter _ to give a new temporary DataFrame with
the correct values. Assign the correct column names to this temporary
DataFrame.

```python
df1
```

```python
# 1. Creating a frm dataframe
frm = pd.DataFrame(df1['From_To'].apply(lambda x:x.split('_')[0]))
frm = frm.rename(columns={'From_To':'From'})
frm
```

```python
# 2. Creating a to dataframe
to = pd.DataFrame(df1['From_To'].apply(lambda x:x.split('_')[1]))
to = to.rename(columns={'From_To':'To'})
to
```

# Q3 
Notice how the capitalisation of the city names is all mixed up in this
temporary DataFrame. Standardise the strings so that only the first letter is
uppercase (e.g. "londON" should become "London".)

```python
# 1. Making the letters well aligned in terms of capitalization in frm dataframe
frm['From'] = frm['From'].apply(lambda x:x.capitalize())
frm
```

```python
# 2. Making the letters well aligned in terms of capitalization in to dataframe
to['To'] = to['To'].apply(lambda x:x.capitalize())
to
```

# Q4 
Delete the From_To column from df and attach the temporary DataFrame from the previous questions.

```python
# 1. Deleting the From_To Series from dataset
df1 = df1.drop('From_To',axis=1)
df1
```

```python
# 2. Concatinating the frm and to datadrames with the main dataframe
df1 = pd.concat([df1,frm,to],axis=1)
df1
```

# Q5
In the RecentDelays column, the values have been entered into the
DataFrame as a list. We would like each first value in its own column, each
second value in its own column, and so on. If there isn't an Nth value, the value
should be NaN.

Expand the Series of lists into a DataFrame named delays, rename the columns
delay_1, delay_2, etc. and replace the unwanted RecentDelays column in df
with delays.

```python
df1
```

```python
# 1. Making the list inside the series an entirely new dataframe
delay_df = df1['RecentDelays'].apply(pd.Series)
delay_df.columns = [f"Delay_{x}" for x in range(1,len(delay_df.columns.tolist()) + 1)]
delay_df
```

```python
# 2. Concatinating the df1 dataframe with the delays dataframe
final = pd.concat([df1,delay_df],axis=1)
final
```

```python
# 3. Renaming the RecentDelays field to Delays
final.rename(columns={'RecentDelays':'Delays'})
```
